# EDA on Centuries of Ultra-Marathon Running Data - 1798-2022

```python
# import relevant libraries

import pandas as pd
import seaborn as sns
```

```python
# CSV dataset imported after downloading from Kaggle - The dataset is from the year 1798 to 2022 (~7M rows)

df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv',low_memory=False)
```

```python
df.head(50)
```

```python
# Cleaning up the data
# Only want relevant data - USA races, 50km/50Mi in 2020-2022
```

```python
# Show 50km or 50Mi
# 50km
# 50mi
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
# Combine 50km/50mi with isin
```

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
# Extract the events happened in 2020 - 2022

df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'].isin([2020,2021,2022]))]
```

```python
# Check for USA specific events

df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'].isin([2020,2021,2022])) & (df['Event name'].str.contains('USA'))]
```

```python
df2.head(10)
```

```python
# Remove USA from Event name as we have already filtered out all USA specific events
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2
```

```python
df2
```

```python
# Clean up athlete age
```

```python
df2['Athlete_Age'] = 2020 - df2['Athlete year of birth']
```

```python
# Remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.drop('Athlete_performance',axis=1,inplace=True)
```

```python
df2
```

```python
# Time to drop columns - Athlete club, Athlete country, Athlete year of birth, Athlete age category
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'],axis=1)
```

```python
df2
```

```python
# Check and Clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['Athlete_Age'].isna()]
```

```python
df2 = df2.dropna()
```

```python
df2
```

```python
# Check for duplicate values
```

```python
df2[df2.duplicated()==True]
```

```python
# Reset the index
```

```python
df2.reset_index(drop=True)
```

```python
# Fix data types
```

```python
df2.dtypes
```

```python
df2['Athlete_Age'] = df2['Athlete_Age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# Rename Columns

# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed         object
# Athlete ID                     int64
# Athlete_Age                  float64
# dtype: object
```

```python
df2 = df2.rename({'Year of event':'year',
                  'Event dates':'race_day',
                  'Event name':'race_name',
                  'Event distance/length':'race_length',
                  'Event number of finishers':'race_number_of_finishers',
                  'Athlete performance':'athlete_performance',
                  'Athlete gender':'athlete_gender',
                  'Athlete average speed':'athlete_average_speed',
                  'Athlete ID':'athlete_id',
                  'Athlete_Age':'athlete_age'},axis=1)
```

```python
df2
```

```python
# Reorder columns
```

```python
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
df3
```

```python
df4 = df3[df3['athlete_gender'] !='X']
```

```python
df4.query('race_length == "50mi" & year == 2020').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>19').head(15)
```

```python
# Seasons for the data --> Slower in summer than winter?

# Spring 3-5
# Summer 6-8
# Fall 9-11
# Winter 12-2

# Split between two decimals
```

```python
df4['race_month'] = df4['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df4['race_season'] = df4['race_month'].apply(lambda x: 'Winter' if x > 11
                                            else 'Fall' if x > 8
                                            else 'Summer' if x > 5
                                            else 'Spring' if x > 2
                                            else 'Winter')
```

```python
df4
```

```python
df4.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
